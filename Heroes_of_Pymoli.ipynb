## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count

gender_purchase_count = purchase_data["Gender"].value_counts()
total_purchases = gender_purchase_count.sum()

female_purchase = gender_purchase_count["Female"]
male_purchase = gender_purchase_count["Male"]
other_nondisclosed_purchase = gender_purchase_count["Other / Non-Disclosed"]

#Average Purchase Price
female_average_price = purchase_data.loc[purchase_data["Gender"] == "Female"]
male_average_price = purchase_data.loc[purchase_data["Gender"] == "Male"]
other_nondisclosed_average_price = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]

updated_female_average_price = female_average_price["Price"].mean()
updated_male_average_price = male_average_price["Price"].mean()
updated_other_nondisclosed_average_price = other_nondisclosed_average_price["Price"].mean()

#Total Purchase Value
female_total_value = purchase_data.loc[purchase_data["Gender"] == "Female"]
male_total_value = purchase_data.loc[purchase_data["Gender"] == "Male"]
other_nondisclosed_total_value = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]

updated_female_total_value = female_total_value["Price"].sum()
updated_male_total_value = male_total_value["Price"].sum()
updated_other_nondisclosed_total_value = other_nondisclosed_total_value["Price"].sum()

#Average Purchase Total Per Person
female_average_purchase_total = updated_female_total_value / female_purchase
male_average_purchase_total = updated_male_total_value / male_purchase
other_nondisclosed_average_purchase_total = updated_other_nondisclosed_total_value / other_nondisclosed_purchase

purchase_count = {
    "Female": female_purchase,
    "Male": male_purchase,
    "Other/Non-Disclosed": other_nondisclosed_purchase
}
average_purchase_price = {
    "Female": updated_female_average_price,
    "Male": updated_male_average_price,
    "Other/Non-Disclosed": updated_other_nondisclosed_average_price
}
total_purchase_value = {
    "Female": updated_female_total_value,
    "Male": updated_male_total_value,
    "Other/Non-Disclosed": updated_other_nondisclosed_total_value    
}
average_purchase_total = {
    "Female": female_average_purchase_total,
    "Male": male_average_purchase_total,
    "Other/Non-Disclosed": other_nondisclosed_average_purchase_total
}

#Summary Data Frame
summary_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": average_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Average Purchase Total Per Person": average_purchase_total
})

#Formatting
summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:,.2f}".format)
summary_df["Total Purchase Value"] = summary_df["Total Purchase Value"].map("${:,.2f}".format)
summary_df["Average Purchase Total Per Person"] = summary_df["Average Purchase Total Per Person"].map("${:,.2f}".format)
summary_df
```
